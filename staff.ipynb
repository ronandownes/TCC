```python
import pandas as pd

import pdfplumber
import pandas as pd

# Open the PDF file
with pdfplumber.open('staff.pdf') as pdf:
    # Initialize empty lists to store names and subjects
    names = []
    subjects = []

    # Process each page of the PDF
    for page in pdf.pages:
        text = page.extract_text()
        
        # Process each line of text to extract names and subjects
        for line in text.split('\n'):
            parts = line.strip().split(" - ")
            if len(parts) == 2:
                names.append(parts[0])
                subjects.append(parts[1])
            elif len(parts) > 2:
                # If there are more than two parts, combine the first parts as the name
                names.append(" - ".join(parts[:-1]))
                subjects.append(parts[-1])

# Create a pandas DataFrame
data = {'Name': names, 'Subject': subjects}
df = pd.DataFrame(data)
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)


# Print the resulting DataFrame


df
```

```python

# Filter the DataFrame
maths_df = df[df['Subject'].str.contains('maths', case=False)]

# Print the filtered DataFrame
maths_df
```
